```python
class Usuario:
    def __init__(self, nombre, usuario, contrase√±a):
        self.nombre = nombre
        self.usuario = usuario
        self.contrase√±a = contrase√±a
    def iniciar_sesion(self, usuario, contrase√±a):
        return self.usuario == usuario and self.contrase√±a == contrase√±a
```

```python
class Empleado(Usuario):
    def __init__(self, nombre, usuario, contrase√±a, cargo):
        super().__init__(nombre, usuario, contrase√±a)
        self.cargo = cargo

    def __str__(self):
        return f"Empleado: {self.nombre} ({self.cargo})"
```

```python
class Cliente:
    def __init__(self, nombre, tipo_cliente="p√∫blico"):
        self.nombre = nombre
        self.tipo_cliente = tipo_cliente

    def mostrar_info(self):
        return f"Cliente: {self.nombre} - Tipo: {self.tipo_cliente}"
```

```python
class Producto:
    def __init__(self, codigo, nombre, precio, cantidad):
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad
```

```python
# ---------- Clase Ferreteria ----------
class Ferreteria:
    def __init__(self, nombre):
        self.nombre = nombre
        self.productos = []
        self.clientes = []

    def agregar_producto(self, producto):
        """Agrega un producto nuevo al inventario."""
        self.productos.append(producto)

    def mostrar_inventario(self):
        """Muestra todos los productos disponibles."""
        print(f"\nüì¶ Inventario de {self.nombre}:")
        if not self.productos:
            print("Inventario vac√≠o.")
        else:
            for p in self.productos:
                print(p)

    def registrar_cliente(self, cliente):
        """Agrega un cliente a la lista de clientes registrados."""
        self.clientes.append(cliente)

    def vender_producto(self, codigo, cantidad, cliente):
        """Procesa una venta y actualiza el stock."""
        for producto in self.productos:
            if producto.codigo == codigo:
                if producto.actualizar_stock(cantidad):
                    total = cantidad * producto.precio
                    print("\n‚úÖ Venta realizada correctamente")
                    print(f"Cliente: {cliente.nombre}")
                    print(f"Producto: {producto.nombre}")
                    print(f"Cantidad: {cantidad}")
                    print(f"Total a pagar: ${total}\n")
                return
        print("‚ùå Producto no encontrado.")
```

```python
# ===============================
# BLOQUE PRINCIPAL DEL PROGRAMA
# ===============================

# Crear empleado del sistema
empleado1 = Empleado("celi", "celi", "clave123", "Vendedora")

print("=== SISTEMA DE FERRETER√çA ===")
print("Inicie sesi√≥n para continuar.\n")

user = input("Usuario: ")
password = input("Contrase√±a: ")

# Verificamos si el usuario y la contrase√±a son correctos
if empleado1.iniciar_sesion(user, password):
    print(f"\nBienvenida {empleado1.nombre} üëã")

    # Crear la ferreter√≠a
    ferreteria = Ferreteria("Ferreter√≠a Central")

    # Cargar productos de ejemplo
    ferreteria.agregar_producto(Producto("A001", "Martillo", 5000, 10))
    ferreteria.agregar_producto(Producto("A002", "Clavos 1 pulgada", 100, 200))
    ferreteria.agregar_producto(Producto("A003", "Taladro", 25000, 5))

    print("\n‚úÖ Productos cargados correctamente en el inventario inicial.")

    # Men√∫ principal
    while True:
        print("\n=== MEN√ö PRINCIPAL ===")
        print("1. Ver inventario")
        print("2. Registrar cliente")
        print("3. Realizar venta")
        print("4. Salir")

        opcion = input("Seleccione una opci√≥n: ")

        if opcion == "1":
            ferreteria.mostrar_inventario()

        elif opcion == "2":
            nombre_cliente = input("Nombre del cliente: ")
            tipo = input("Tipo de cliente (registrado/p√∫blico): ")
            nuevo_cliente = Cliente(nombre_cliente, tipo)
            ferreteria.registrar_cliente(nuevo_cliente)
            print("‚úÖ Cliente registrado correctamente.")

        elif opcion == "3":
            ferreteria.mostrar_inventario()
            codigo = input("Ingrese el c√≥digo del producto: ")
            try:
                cantidad = int(input("Cantidad a vender: "))
            except ValueError:
                print("‚ùå Ingrese un n√∫mero v√°lido.")
                continue

            nombre_cliente = input("Nombre del cliente: ")
            cliente_temp = Cliente(nombre_cliente)
            ferreteria.vender_producto(codigo, cantidad, cliente_temp)

        elif opcion == "4":
            print("üëã Cerrando sesi√≥n. ¬°Hasta luego!")
            break

        else:
            print("‚ùå Opci√≥n no v√°lida. Intente de nuevo.")
else:
    print("\n‚ùå Usuario o contrase√±a incorrectos.")
```
